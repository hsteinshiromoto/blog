```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import matplotlib 
# matplotlib.rc('xtick', labelsize=20) 
# matplotlib.rc('ytick', labelsize=20) 
params = {'axes.labelsize': 16 #Fontsize of the x and y labels
         ,'axes.titlesize': 16 #Fontsize of the axes title
         ,'figure.titlesize': 16 #Size of the figure title (Figure.suptitle())
         ,'xtick.labelsize': 16 #Fontsize of the tick labels
         ,'ytick.labelsize': 16 #Fontsize of the tick labels
         ,'legend.fontsize': 16 #Fontsize for legends (plt.legend(), fig.legend())
         ,'legend.title_fontsize': 16
         }

plt.rcParams.update(params)
```

```python
x = np.arange(-2, 2, 0.01)
y = x
```

```python
f = lambda x,y: np.maximum(x,y)
g = lambda x,y: np.log(np.exp(x) + np.exp(y))
```

```python
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
W = g(X, Y)
```

```python
fig = plt.figure(figsize=(20, 10))
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

```python
fig = plt.figure(figsize=(20, 10))
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, W, 50)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

```python
fig = plt.figure(figsize=(20, 10))
ax = plt.axes(projection='3d')
ax.plot_wireframe(X, Y, Z)
ax.set_title('wireframe');
```

```python
fig = plt.figure(figsize=(20, 10))
ax = plt.axes(projection='3d')
ax.plot_wireframe(X, Y, W)
ax.set_title('wireframe');
```

```python
fig = plt.figure(figsize=(20, 20))
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, linewidth=0, antialiased=False, cmap='coolwarm', edgecolor='black')
ax.set_title('surface')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
fig.tight_layout()
plt.savefig("f.svg")
```

```python
fig = plt.figure(figsize=(20, 20))
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, W, linewidth=0, antialiased=False, cmap='coolwarm', edgecolor='black')
ax.set_title('surface')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
fig.tight_layout()
plt.savefig("g.svg")
```

```python
h = g(x,0) - f(x,0)
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
ax.plot(x, h, label=r"$g-f$")
ax.plot(x, f(x,0), label="f")
ax.plot(x, g(x,0), label="g")
ax.legend()
ax.set_xlabel('x')
ax.set_ylabel('functions')
```
