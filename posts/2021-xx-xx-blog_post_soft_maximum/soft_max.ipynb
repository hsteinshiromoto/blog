```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 2, 0.01)
y = x
```

```python
f = lambda x,y: np.maximum(x,y)
g = lambda x,y: np.log(np.exp(x) + np.exp(y))
```

```python
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
W = g(X, Y)
```

```python
fig = plt.figure(figsize=(20, 10))
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

```python
fig = plt.figure(figsize=(20, 10))
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, W, 50)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

```python
fig = plt.figure(figsize=(20, 10))
ax = plt.axes(projection='3d')
ax.plot_wireframe(X, Y, Z)
ax.set_title('wireframe');
```

```python
fig = plt.figure(figsize=(20, 10))
ax = plt.axes(projection='3d')
ax.plot_wireframe(X, Y, W)
ax.set_title('wireframe');
```

```python
fig = plt.figure(figsize=(20, 20))
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, linewidth=0, antialiased=False, cmap='coolwarm', edgecolor='black')
ax.set_title('surface')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.savefig("test.svg")
```

```python
fig = plt.figure(figsize=(20, 10))
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, W, rstride=1, cstride=1, cmap='viridis', edgecolor='none')
ax.set_title('surface')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.savefig("test.svg")
```
