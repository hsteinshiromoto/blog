```python
import numpy as np
import matplotlib.pyplot as plt
import xkcd
```

```python
t = np.arange(0, 10)
```

```python
with plt.xkcd():
    plt.plot(t, np.exp(-t))
    plt.xlabel("Weeks Since 1st of January")
    plt.title("Proportion of People Saying \"Happy New Year\"")
    plt.tight_layout(pad=0.5)
    plt.savefig('hny.png')
```
